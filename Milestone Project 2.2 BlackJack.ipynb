```python
import random

##_____________________________THE DECK(GLOBAL VARs)____________________________
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

##____________________________CLASS________________________________

class Card():
    def __init__(self,suit,rank): #Creates a Card object
        self.suit = suit
        self.rank = rank
        
    def __str__(self): #Print the Card
        return self.rank + " of " + self.suit
    
class Deck:
    def __init__(self):
        self.deck = [] #Makes a new list of Cards 
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank)) ##Makes a self.deck with Card class for each card
        
    def __str__(self): 
        deck_comp = '' #What the deck contains
        for card in self.deck:
            deck_comp += '\n' + card.__str__() #Prints each card.__str__ in a new line
        return "The deck has: " +deck_comp
    
    def shuffle(self):
        random.shuffle(self.deck) #Shuffles the deck (No return)
        
    def deal(self):
        single_card = self.deck.pop() #Holds a poped Deck card and return it
        return single_card
    
class Chips:
    def __init__(self):
        self.total = 100
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet

class Hand():
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card): #card is called from Deck.deal()##
        self.cards.append(card) #add that "card" to the self.cards
        self.value += values[card.rank] #set self.value to add card value from the dictionary Values
        if card.rank == "Ace": #Track aces
            self.aces += 1
            
    def adjust_for_aces(self): #(1)if value is over 21 and i have aces over 0
        while self.value > 21 and self.aces: #(2) then decrease from value 10
            self.value -= 10 #(3) and decrease aces by 1
            self.aces -=1
            
##____________________________FUNCTIONS______________________
def take_bet(chips):
    while True:
            try:
                chips.bet = int(input("How many chips you want bet?"))
            except:
                print("Please enter an Integer")
            else:
                if chips.bet > chips.total:
                    print(f"Not enough chips! You only have: {chips.total}")
                else:
                    break
                    
def hit(deck,hand):
    single_card = deck.deal()
    hand.add_card(single_card)
    hand.adjust_for_aces()
    
def hit_or_stand(deck,hand):
    global playing
    while True:
        x = input("Hit or Stand? h/s")
        if x[0].lower() == "h":
            hit(deck,hand)
            show_some(player_hand,dealer_hand)
        elif x[0] == 's':
            print("Player Stands,Dealer Turn")
            playing = False
            break
        else:
            print("Enter a correct value 's' or 'h' ")
            continue

def player_busts(player,dealer,chips):
    print("BUSTED PLAYER")
    chips.lose_bet()
def player_wins(player,dealer,chips):
    print("PLAYER WIN")
    chips.win_bet()
def dealer_busts(player,dealer,chips):
    print("PLAYER Wins Delaer BUSTED")
    chips.win_bet()
def dealer_wins(player,dealer,chips):
    print("Dealer WINS")
    chips.lose_bet()
def push(player,dealer):
    print("Dealer and Player tie!")

def show_some(player,dealer):
    print("Dealers Hand:")
    print("Hidden card")
    print(dealer.cards[1])
    print("\n")
    print("Players Hand:")
    for card in player.cards:
        print(card)
        
def show_all(player,dealer):
    print("Dealers Hand:")
    for card in dealer.cards:
        print(card)
    print("\n")
    print("Players Hand:")
    for card in player.cards:
        print(card)
    
playing = True #Game status

##______________________GAME ON_________________
while True:
    print("WELCOME TO BLACK-JACK") #First Statment
    
    #Create and shuffle a Deck and deal 2 cards to each player
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand() #Player + 2 cards
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand = Hand() #Dealer + 2 cards
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    #Set up the player's chips
    player_chips = Chips()
    
    #Take players bet
    take_bet(player_chips)
    
    #show Cards with hidden Dealer card
    show_some(player_hand,dealer_hand)
    
    while playing:
        #Ask player to hit or stand
        hit_or_stand(deck,player_hand)
        
        #show cards with hidden delaer card
        show_some(player_hand,dealer_hand)
        
        #Check if player over 21 run player_bust() and break
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            break
            
    #if player hasnt busted , play delaer until he reaches 17
    if player_hand.value <= 21:
        while dealer_hand.value <= 17:
            hit(deck,dealer_hand)
         
        show_all(player_hand,dealer_hand)
        
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,chips,player_chips)
        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,chips,player_chips)
        else:
            push(player_hand,dealer_hand)
            
    print(f"players total chips are at: {player_chips.total}")
    
    new_game = input("new game?")
    if new_game[0].lower() == 'y':
        playing = True
        continue
    else:
        print("GoodBye")
        break
```
