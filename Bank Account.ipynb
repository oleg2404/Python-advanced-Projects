```python
'''
Created on 11 Oct 2019

@author: Oleg Ovroutsky
'''
import random

accounts = {} ##Global dictionary holder for bank accounts by name of the account
pins = [] ## Global holder for pin codes in the system
'''
3 Classes: main class of account and 3 account types
Checking account
Savings account
Bussiness account
'''
class Account: ##Account class Father class
    
    def __init__(self,owner):
        self.owner = owner
        print("Account was created successfully")
        
    def __str__(self):
        print(f"{self.owner}")
        
class CheckAccount(Account): ##Checking account class
    
    def __init__(self,owner,pin,first_depo = 0): ## Takes in owners name,pin code for login and first deposit
        self.pin = pin
        Account.__init__(self,owner)
        self.balance = int(first_depo)
        print(f"Welcome Aboard {self.owner}")

    def withdraw(self,amount):
        if amount > self.balance:
            print(f"Not enough Money, Available: {self.balance}")
        else:
            self.balance -= int(amount)
            print(f"{amount} was withdrawn")
    
    def deposit(self,amount):
        self.balance += int(amount)
        return f"{amount} was deposited to your account {self.owner}"

    def get_balance(self):
        return self.balance
    
    def __str__(self):
        super().__str__()
        return f"\nYour Balance: {self.balance}"
    
class SavingsAccount(Account):## Savings account class
    
    def __init__(self,owner,pin,first_depo = 0):## takes in owners name,code for login,and first depo
        self.pin = pin
        Account.__init__(self,owner)
        self.balance = int(first_depo)
        print(f"Your Savings Account was created {self.owner}")
        
    def withdraw(self,amount):
        
        if amount > self.balance:
            print(f"Not enough Money, Available: {self.balance}")
        else:
            self.balance -= int(amount)
            print(f"{amount} was withdrawn")
    
    def deposit(self,amount):
        self.balance += int(amount)
        return f"{amount} was deposited to your savings account {self.owner}"

    def get_balance(self):
        return self.balance
    
    def __str__(self):
        super().__str__()
        return f"\nYour Savings Account Balance is: {self.balance}"

class BusAccount(Account):##Business acoount class
    
    def __init__(self,owner,pin,first_depo = 0):##Takes in owners name,code for login,first deposit
        self.pin = pin
        Account.__init__(self,owner)
        self.balance = int(first_depo)
        print(f"Your Business Account was created {self.owner}")        
    def withdraw(self,amount):
        if amount > self.balance:
            print(f"Not enough Money, Available: {self.balance}")
        else:
            self.balance -= int(amount)
            print(f"{amount} was withdrawn")
    
    def deposit(self,amount):
        self.balance += int(amount)
        return f"{amount} was deposited to your Business account {self.owner}"

    def get_balance(self):
        return self.balance
    
    def __str__(self):
        super().__str__()
        return f"\nYour Business Account Balance is: {self.balance}"

def sign_up(): ##Function to open a new account and store in in the accounts dictionary and pin codes in pins list
    
    global pins
    global accounts
    
    print("Welcome\nWich Account type would you like to open?\n1.Checking Account\n2.Bussiness Account\n3.Savings Account")
    
    while True:
        try:
            answer = int(input("1/2/3 ?"))
        except:
            print("Not a valid answer!!")
        else:
            break
            
    name = input("Enter your Name")
    
    while True:
        pin = random.choice(range(1000,9999))
        if pin in pins:
            continue
        else:
            pins.append(pin)
            break
            
    deposit = int(input("How much Money would you like to Deposit?"))
    
    if answer == 1:
        accounts[name] = CheckAccount(name,pin,deposit)
        print(f"Your pin code is: {accounts[name].pin}")
    if answer == 2:
        accounts[name] = BusAccount(name,pin,deposit)
        print(f"Your pin code is: {accounts[name].pin}")
    if answer == 3:
        accounts[name] = SavingsAccount(name,pin,deposit)
        print(f"Your pin code is: {accounts[name].pin}")
        
        
def log_in(): ##Log in to your account by pin code that is unique to your account from accounts dictionary and pins list
    global pins
    global accounts
    print("Welcome to the ATM\nPlease Enter your PIN Number:(****)")
    
    while True:
        pin = int(input("PIN"))
        if pin in pins:
            for account in accounts:
                if accounts[account].pin == pin:
                    while True:
                        print(accounts[account])
                        answer = int(input("What action would you like to take?\n1.withdraw\n2.Deposit\n3.Exit"))
                        if answer == 1:
                            while True:
                                answer = int(input("How Much to Withdraw?"))
                                if (accounts[account].balance - answer) < 0:
                                    print("Not enough money")
                                    continue
                                accounts[account].withdraw(answer)
                                break
                        if answer == 2:
                            answer = int(input("How much to Deposit?"))
                            accounts[account].deposit(answer)
                        if answer == 3:
                            break
            break
        else:
            print("Invalid PIN CODE")
            break
def atm(): ##Function that runs the ATM MACHINE
    while True:
        answer = int(input("Hello.\nWhat would you like to do?\n1.Open a new Account?\n2.Log in to your account?\n3.Exit\n1/2/3?"))
        if answer == 1:
            sign_up()
        if answer == 2:
            log_in()
        if answer == 3:
            print("Good Day")
            break
        
```
